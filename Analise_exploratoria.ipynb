### **Correlação do gênero com o salário anual**

```python
full_data_2[['Gender','Yearly brutto salary (without bonus and stocks) in EUR']].groupby('Gender').describe().round(1)
```

É possível observar que o salário bruto anual médio foi mais alto para os funcionários do sexo masculino. Além disso, a mediana de salários para os homens foi de 70000 euros e para as mulheres foi de 60000 euros. Os valores de salários máximos observados são iguais para ambos os sexos. 

### **Correlação do gênero com os valores de bônus e ações**

```python
full_data_2[['Gender','Yearly bonus + stocks in EUR']].groupby('Gender').describe().round(1)
```

É possível observar que a média de bonus e ações foi mais alta para os funcionários do sexo masculino. Além disso 25% dos maiores valores obtidos estão acima de 5000 euros anuais para as mulheres e 10000 euros anuais para os homens.

### **Correlação dos anos de experiencia com o salário anual**

```python
full_data_2[['Total years of experience','Yearly brutto salary (without bonus and stocks) in EUR']].groupby('Total years of experience').describe().round(1)
```

```python
plt.figure(figsize = (27,10))
sns.regplot(x='Total years of experience', y='Yearly brutto salary (without bonus and stocks) in EUR', data=full_data_2);
```

A partir do grafico acima descrito é possível notar uma tendencia de aumento do salário anual em relação aos anos de experiência. 

### **Correlação dos anos de experiencia com o cargo**

```python
full_data_2[['Position ','Total years of experience']].groupby('Position ').describe().round(1)
```

```python
plt.figure(figsize = (27,10))
sns.barplot(x='Position ', y='Total years of experience', data=full_data_2);
```

É possivel observar no gráfico e na tabela acima gerados que o cargo de gerente (Manager) foi o que apresentou maiores valores de tempo de experiencia, seguido do cargo de Engenheiro de Software.

### **Correlação do tamanho da empresa com a quantidade de dias de férias**
